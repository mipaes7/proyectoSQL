```python
import pandas as pd

# Cargar los datasets
datas = pd.read_csv('data/clase_1.csv', delimiter=';')
datas2 = pd.read_csv('data/clase_2.csv', delimiter=';')
fstack = pd.read_csv('data/clase_3.csv', delimiter=';')
fstack2 = pd.read_csv('data/clase_4.csv', delimiter=';')


# Combinar datasets
dataclass = pd.concat([datas, datas2], axis=0).reset_index(drop=True)
fstackclass = pd.concat([fstack, fstack2], axis=0).reset_index(drop=True)

# Concatenar todos los estudiantes y añadir columna ID
estudiantes = pd.concat([dataclass, fstackclass], axis=0).reset_index(drop=True)
estudiantes['id_estudiante'] = range(1, len(estudiantes) + 1)

# Cambiar el nombre de la columna Proyecto_FullSatck a Proyecto_FullStack
estudiantes = estudiantes.rename(columns={'Proyecto_FullSatck': 'Proyecto_FullStack'})

# Seleccionar las columnas necesarias para el DataFrame de estudiantes
df_estudiantes = estudiantes[['id_estudiante', 'Nombre', 'Email', 'Promoción', 'Fecha_comienzo', 'Campus']]

# Mostrar el DataFrame de estudiantes
df_estudiantes
```

```python
# Obtener nombres de los proyectos
proyectos = estudiantes.columns[5:-1]

# Crear un DataFrame de proyectos con id_proyecto
df_proyectos = pd.DataFrame({'Nombre_proyecto': proyectos})
df_proyectos['id_proyecto'] = range(1, len(proyectos) + 1)
df_proyectos
```

```python
aptitudes=estudiantes.iloc[:,5:]
aptitudes

data_proyectos = {
    'id_proyecto': range(1, 11),
    'Nombre_proyecto': ['Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 'Proyecto_ML', 'Proyecto_Deployment', 'Proyecto_WebDev', 'Proyecto_FrontEnd', 'Proyecto_Backend', 'Proyecto_React', 'Proyecto_FullStack']
}

proyectos = pd.DataFrame(data_proyectos)


# Paso 1: Transformar aptitudes a formato largo
aptitudes_largo = aptitudes.melt(id_vars=['id_estudiante'], 
                                 var_name='Nombre_proyecto', 
                                 value_name='Status')

# Paso 2: Filtrar filas con NaN en 'Status'
aptitudes_largo = aptitudes_largo.dropna(subset=['Status'])

# Paso 3: Mapear 'Nombre_proyecto' a 'id_proyecto'
aptitudes_final = aptitudes_largo.merge(proyectos, on='Nombre_proyecto', how='left')

# Seleccionar y reordenar las columnas necesarias
aptitudes_final = aptitudes_final[['id_estudiante', 'id_proyecto', 'Status']]

# Mostrar el resultado final

aptitudes_final.sort_values(by='id_estudiante',inplace=True)
aptitudes_final.reset_index(drop=True,inplace=True)
print(aptitudes_final[aptitudes_final['id_estudiante']==4])

print(estudiantes[estudiantes['id_estudiante']==4])

aptitudes_final
aptitudes_final['aptitud_id'] = range(1, len(aptitudes_final) + 1)
aptitudes_final
```

```python



# Crear DataFrame campus
campus = estudiantes[['Campus', 'id_estudiante']].drop_duplicates().reset_index(drop=True)
campus['Campus_id'] = campus.index + 1

# Definir una función para asignar Campus_id según el nombre del campus
def asignar_campus_id(campus_name):
    if campus_name == 'Madrid':
        return 1
    elif campus_name == 'Valencia':
        return 2
    else:
        return None  # Manejar cualquier otro caso según sea necesario

campus
# Crear DataFrame campus sin la columna id_estudiante
campus = estudiantes[['Campus']].drop_duplicates().reset_index(drop=True)

# Definir una función para asignar Campus_id según el nombre del campus
def asignar_campus_id(campus_name):
    if campus_name == 'Madrid':
        return 1
    elif campus_name == 'Valencia':
        return 2
    else:
        return None  # Manejar cualquier otro caso según sea necesario

# Aplicar la función a la columna 'Campus' para obtener 'Campus_id'
campus['Campus_id'] = campus['Campus'].apply(lambda x: asignar_campus_id(x))

# Mostrar el DataFrame campus resultante
campus

```
